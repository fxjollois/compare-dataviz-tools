# Python - `matplotlib`

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import pandas

%matplotlib inline

temp = pandas.read_csv("../HadCRUT4-gl-dowload20170601.csv", sep = ";")
#temp.head()
```

```python
temp = temp.assign(Min = [min(temp[temp.columns[1:13]].loc[i]) for i in range(temp.shape[0])])
temp = temp.assign(Max = [max(temp[temp.columns[1:13]].loc[i]) for i in range(temp.shape[0])])
# temp.head()
```

```python
plt.figure(figsize = (16, 9))
plt.plot("Year", "Annual", data = temp, color = "steelblue")
plt.annotate(xy = (1961, min(temp.Min)), s = "Reference period")
plt.fill_between("Year", "Min", "Max", data = temp, color = "lightblue", zorder = 2)
plt.ylabel("Temperature anomaly")
plt.hlines(0, xmin = min(temp.Year), xmax = max(temp.Year), colors = "#BBBBBB", linestyles = "dashed")
plt.gca().add_patch(Rectangle((1961, -1), 1990 - 1961, 2, color = "#EEEEEE"))
```
