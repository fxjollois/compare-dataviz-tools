# Python - `bokeh`

```python
from bokeh.plotting import figure
from bokeh.charts import Area, Line, output_notebook, show
from bokeh.models import Plot, BoxAnnotation, Label, Span, ColumnDataSource, DataRange1d
from bokeh.models.glyphs import Patch

import numpy
import pandas
output_notebook()
```

```python
temp = pandas.read_csv("../HadCRUT4-gl-dowload20170601.csv", sep = ";")
temp = temp.assign(Min = [min(temp[temp.columns[1:13]].loc[i]) for i in range(temp.shape[0])],
                   Max = [max(temp[temp.columns[1:13]].loc[i]) for i in range(temp.shape[0])])
# temp.head()
```

```python
p = figure(plot_width = 800, plot_height = 500, y_axis_label = "Temperature anomaly")

p.add_layout(Span(location = 0, dimension = 'width', line_color='#999999', line_dash='dashed'))
p.add_layout(BoxAnnotation(left = 1961, right = 1990, fill_color = '#dddddd'))
p.add_layout(Label(x = 1961, y = min(temp.Min), text = 'Reference period', text_font_size = "10pt"))

source = ColumnDataSource(dict(x = numpy.append(temp.Year, temp.Year[::-1]), 
                               y = numpy.append(temp.Min, temp.Max[::-1])))
glyph = Patch(x="x", y="y", fill_color="lightblue", line_color = "lightblue")
p.add_glyph(source, glyph)

p.line(temp.Year, y = temp.Annual, color = "steelblue")


show(p)
```
