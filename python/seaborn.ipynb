# Python - `seaborn`

```python
import seaborn
import matplotlib.pyplot as plt
import pandas

%matplotlib inline

temp = pandas.read_csv("../HadCRUT4-gl-dowload20170601.csv", sep = ";")
#temp.head()
```

```python
temp = temp.assign(Min = [min(temp[temp.columns[1:13]].loc[i]) for i in range(temp.shape[0])],
                   Max = [max(temp[temp.columns[1:13]].loc[i]) for i in range(temp.shape[0])])
# temp.head()
```

```python
seaborn.set_style("white")
fig, g = plt.subplots(figsize = (16, 9))
p = seaborn.tsplot(temp.Annual, time = temp.Year)
p.set_ylabel("Temperature anomaly")
p.axhline(y = 0, color = "#BBBBBB", zorder = -1)
p.annotate(s = "Reference period", xy = (1961, min(temp.Min)))
p.axvspan(xmin = 1961, xmax = 1990, ymin = -1, ymax = 1, facecolor = "#DDDDDD", alpha = .5, zorder = -2)
p.fill_between("Year", "Min", "Max", data = temp, color = "lightblue", zorder = 2)
```
